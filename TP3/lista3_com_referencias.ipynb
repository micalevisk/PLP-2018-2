Neste sentido, a linguagem é declarativa, sem efeitos colaterais. #### Questão 1-2: Por que predicados como os citados acima podem ser considerados indesejáveis?

Predicados como `assert` e `retract` proporcionam dinamicidade na base de conhecimentos, esssa natureza dinâmica do código tende a impactar negativamente a legibilidade do programa, já que, quanto mais dinâmico, mais difícil será controlar o que realmente pode ou não ser consultado (manter um estado mental). Além de gerar efeitos colaterais. Outros predicados de alta-ordem (ou meta-predicados), como `call`, colaboram com esses efeitos.

 #### Questão 1-3: Como variáveis em Prolog são diferentes de variáveis nas outras linguagens que temos visto neste curso, como C, C++, Java e Pyhton?

As variáveis do Prolog, em termos de sintaxe, devem ser definidas de uma maneira específica: começar com uma letra maiúscula ou com um *underscore* para serem tratadas como tal. Além diso, a variável **"`_`"** é chamada de _variável anônima_, por ter um significado especial atrelado ao _pattern matching_ proposto pela linguagem. Uma vez atribuída, uma variável não pode ser sobrescrita.

Elas possuem 3 funções principais:
1. Denotar objetos específicos porém desconhecidos;
2. Servir de *place holders* para assegurar a aridade de um símbolo funcional ou de um predicado;
3. Agir como transportadores de dados ou como correferência de *constraints*.

No primeiro caso, as variáveis podem ser usadas para denotar um objeto cujas propriedades estão especificadas de uma maneira onde não especificam o objeto em si. Uma classe de objetos pode ser especificada de uma maneira implícita.

No segundo caso, a aridade dos símbolos, ou o número constante de argumentos, não pode ser mudado.

No terceiro caso, duas ou mais ocorrências de uma variável numa expressão denotam o mesmo objeto, e se qualquer mudança ocorrer a alguma dessas variáveis, todos as variáveis devem ser substituídas pelo mesmo símbolo ou valor.

No Prolog, a atribuição das variáveis é feita de maneira diferente.  Variáveis podem ser atribuídas de duas formas:
1. onde a atribuição se utilizando do símbolo **"`=`"** é feita para confinar uma variável a um valor, não podendo se sobreescrever outro valor a esta variável
2. Já o símbolo **"`is`"** é tido como uma atribuição no senso mais comum visto em outras linguagens.

##### Solução inicial; mais intuitiva:

````python
%% percurso(?Origem, ?Destino, ?Distancia).
percurso(novayork, londres, 10).
percurso(londres, paris, 5).
percurso(moscou, paris, 8).
percurso(paris, roma, 3).
percurso(japao, sidney, 8).
percurso(japao, pequim, 10).
percurso(sidney, roma, 12).
percurso(pequim, alexandria, 8).
percurso(alexandria, roma, 5).


%% esta_no_final(+Val, +Lin, -ResultList)
%% True se `ResultList` tem o elemento `Val` inserido no final da lista `Lin`
esta_no_final(Val, Lin, ResultList) :- append(Lin, [Val], ResultList).

%% caminho(+Origem, +Destino, -Caminho, -DistanciaPercorrida)
%% True se existe um `Caminho` com `DistanciaPercorrida` entre `Origem` e `Destino`
caminho(Origem, Destino, Caminho, DistanciaPercorrida) :- %% predicado auxiliar
    caminho_(Origem, Destino, [], Caminho, DistanciaPercorrida).

caminho_(X, Y, PercursoAnterior, PercursoXY, DistanciaXY) :-
    percurso(X, Y, DistanciaXY),
    esta_no_final(percurso(X, Y), PercursoAnterior, PercursoXY).

caminho_(X, Y, PercursoAnterior, PercursoXY, DistanciaXY) :-
    percurso(X, D, DistanciaXD),
    esta_no_final(percurso(X, D), PercursoAnterior, PercursoXD),
    caminho_(D, Y, PercursoXD, PercursoXY, DistanciaDY),
    DistanciaXY is DistanciaXD + DistanciaDY.
````

```
swap(WSatual, WSnova) :- 
    top(WSatual, N1), pop(WSatual, WS2), 
    top(WS2, N2), pop(WS2, WS3), 
    push(WS3, N1, WS4),

# Para casos em que não há 2 elementos no topo da pilha, deixe-a inalterada
```

Contudo, uma forma mais elegante em Prolog é a que usa o motor de unificação para casar padrões. Neste caso, _swap_ ficaria assim:

```
swap([N1, N2|WSresto], [N2, N1|WSresto]).
swap(WS, WS). 
```

Baseado em sua experiência com Prolog, a compare com Forth considerando os seguintes critérios:
* Simplicidade, Ortogonalidade, Tipos de Dados, Projeto de Sintaxe, Suporte à Abstração, Expressividade, Checagem de Tipos, Manipulação de Exceções e Restrição de Aliases

| | | Legibilidade | Escrita | Confiabilidade |
|:-:|:-|:------------:|:-------:|:--------------:|
| `1` | Simplicidade                       | **X** | **X** | **X**
| `2` | Ortogonalidade                 | **X** | **X** | **X**

Em relação à **simplicidade**, o Forth se mostra melhor por apresentar apenas uma construção lógica: usar algumas instruções que atuam sobre 2 pilhas. O que gera um grau de **ortogonalidade** maior. Enquanto, apesar da semântica declarativa ajudar na simplicidade, o Prolog introduz "termos", "predicados", "fatos" e "regras", para compor a lógica simbólica da linguagem. Além de fornecer notações _syntax sugar_, que tendem a ajudar problemas específicos mas aumenta o quantidade de formas de soluções.

Em relação ao **projeto de sintaxe**, a forma dos elementos da linguagem Forth prejudica a legibilidade se muitas instruções forem informadas de forma dependente. Enquanto o Prolog tem a vantagem de possibilitar um nível de indentação. Além de fornecer um conjunto mais conhecido de palavras reservadas.

Ao definir **tipos de dados**, o Prolog tem um poder de abstração e legibilidade melhoradas.  E a  **capacidade de abstração** do processo em Forth é mais simples e consistente, o que dificulta a legibilidade se for preciso definir estruturas mais complexas (muitas instruções aninhadas).

A semântica do Prolog está atrelada à sua capacidade de expressão, que, comparada com a do Forth -- no contexto dos domínios para quais foram projetadas, são equivalentes, ainda que o Forth forneça formas mais convenientes no geral. Porém se apresenta mais confiável pois, em Forth, é necessário manter em mente a manipulação dos estados das pilhas para desenvolver uma sequência útil de instruções.

O suporte à checagem de tipos, manipulação de exceção e restrição de _aliases_ não podem ser comparados pois, em Forth, essas características não foram explicitadas.
