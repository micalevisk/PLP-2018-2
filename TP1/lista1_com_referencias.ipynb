#### Questão 1-1: O que caracteriza as quatro primeiras gerações de linguagens de programação? Dê exemplos de linguagens em cada uma destas gerações:  

vide
- https://www.inf.pucrs.br/~gustavo/disciplinas/pli/material/paradigmas-aula01.pdf

- Primeira geração (1995):
  + código executado diretamente pela CPU (via _front panel_)
  + baixa capacidade de abstração
  + instruções usando apenas representação em números binários (0s e 1s)
  + pouco intuitiva; difícil de interpretar e ler, mas apropriadas para máquinas
  + referenciar variável (registrador) com endereço absoluto
  + exemplo: Linguagem de máquina

- Segunda geração (1965):
  + linguagens de baixo nível
  + notação legível para humanos do código de máquina utilizado para programar códigos para microprocessadores ou microcontroladores
  + o código escrito pelo programador deveria ser convertido em um código legível por uma máquina -- processo chamado de "assembly" realizado por montadores
  + linguagem específica a uma família particular de processadores e ambientes
  + linguagens geralmente usadas em _kernel_ e dispositivos de _drivers_ mas adequada para processamento intensivos
  + correspondência 1 por 1 entre o código programado e código gerado (resultante)
  + exemplo: Assembly

- Terceira geração (1972):
  + linguagens de alto-nível
  + independente da plataforma alvo
  + linguagem projetada para ser facilmente lida por um ser humano (expressividade)
  + inclusão de varáveis com nomes e definição de tipo de dados
  + exemplo: Fortran, ALGOL, COBOL, C, Java, BASIC e Pascal

- Quarta geração:
  + constituem uma consolidação e correção de características na geração anterior, porém, com novas facilidades importantes
  + linguagens que permitem ao programador especificar o que deve ser feito visando um resultado imediato
  + suporte para gerenciamento de banco de dados, otimização de operações matemáticas e desenvolvimento web
  + subconjunto de linguagens de domínio-específico (DSLs)
  + exemplo: SQL, XML

#### Questão 1-2: Por que linguagens modernas tendem a suportar múltiplos paradigmas?

vide

Além disso, a popularidade de uma linguagem de programação está relacionada aos conceitos que serão utilizados nela -- flexibilidade. Assim, desenvolvedores experientes em diferentes linguagens e, naturalmente, paradigmas, serão mais facilmente alcançados.

#### Questão 2-1: Defina os seguintes conceitos como são compreendidos no paradigma funcional: (a) funções anônimas, (b) funções de primeira classe, (c) funções de alta ordem, (d) funções recursivas, (e) funções puras, (f) avaliação preguiçosa, (g) iteradores e (h) geradores ou streams.

a. **funções anônimas**:
Funções definidas sem um nome associado. Normalmente usadas como expressões que serão avaliadas apenas uma vez. Diminuindo, assim, o _overhead_ causado pela existência de mais uma função no escopo. Geralmente limitadas a uma única instrução que recebem algo, realiza um processamento simples e  o retorna.

b. **funções de primera classe:**
Quando funções são tratadas como "objetos de primeira-classe". Ou seja, são como qualquer outro tipo de dados, assim, podem ser passadas para outras funções como argumento, retornadas por elas e atribuídas como valor à variáveis.

c. **funções de alta ordem:**
São funções que trabalham com funções de primeira classe, isto é, recebem funções como argumento e/ou retornam funções. Exemplo: `map`, `filter` e `reduce`.

d. **funções recursivas:**
Usado como método de iteração onde uma função é definida usando ela mesma. Na prática, é uma função que chama a si própria até que não seja mais necessário (atingir um caso base).

e. **funções puras:**
Funções que geram uma saída em função unicamente dos dados de entrada. Além de não produzirem efeitos colaterais. Proporciando a Transparência Referencial, que é a capacidade de substituir uma chamada da função pelo seu valor, já que a função terá a mesma saída sempre que forem passados as mesmas entradas.

f. **avaliação preguiçosa:**
Capacidade de avaliar uma expressão o mais tarde possível -- somente quando for realmente necessário. Tornando a modularização vivável, já que é possível construir diversos cenários sem consumir os recursos gerados por todos eles ao mesmo tempo. Além da capacidade de representar potencialmente estruturas de dados infinitas.

g. **iteradores:**
Recurso da Computação Preguiçosa, usados como uma forma de processar uma série/sequência de elementos. Retornam uma sequência de dados que são gerados por demanda, elemento a elemento, as _streams_.

h. **geradores:**
São expressões que criam (geram) iteradores. Exemplo a função _built-in_ do Python2 `xrange`.

````
def soma_quadrados(n):
  somar = lambda x, y: x+y
  quadrado = lambda n: n**2
  return reduce(somar, [ quadrado(n) for n in range(1, n) ])



## --------------------------------------- 
soma_quadrados(5)
````

#### Questão 2-3: números perfeitos
Um número inteiro positivo é _perfeito_ se ele é a soma de todos os seus fatores, excluído o próprio número. Escreva a função `perfeitos(n)` que retorna todos os números perfeitos entre os ***n***  primeiros números inteiros. Ex: `perfeitos(10000) = [6, 28, 496, 8128]`.

````
def perfeitos(n):
  divisivel = lambda n, x: n % x == 0
  fatores = lambda n: ( x for x in range(1, n) if divisivel(n, x) )

  return [
      x for x
      in range(1, n)
      if sum( fatores(x) ) == x
  ]



## --------------------------------------- 
perfeitos(10000)
````

#### Questão 2-4: índices na lista como função de alta ordem
Usando recursão, implemente a função `indices_na_lista(lst, el, comp)` que retorna os índices na lista `lst`, que satisfazem a comparação com o elemento `el`, de acordo com o comparador `comp`.

Exs:
- `indices_na_lista([10, 20, 30, 20, 10], 20, lambda x, y: x == y) == [1, 3]`
- `indices_na_lista([10, 20, 30, 20, 10], 20, lambda x, y: x < y) == [0, 4]`
- `indices_na_lista(['10', 20, '30', '20', 10], '20', lambda x, y: x == y if type(x) == type(y) else False) == [3]`

````
def indices_na_lista(lst, el, comp):
  def iterate_from_index(i, limit):
    return [] if i == limit \
      else ( [i] if comp(lst[i], el) else [] ) + iterate_from_index(i + 1, limit)

  return iterate_from_index(0, len(lst))



## --------------------------------------- 
indices_na_lista([10, 20, 30, 20, 10], 20, lambda x, y: x == y), \
indices_na_lista([10, 20, 30, 20, 10], 20, lambda x, y: x < y),  \
indices_na_lista(['10', 20, '30', '20', 10], '20', lambda x, y: x == y if type(x) == type(y) else False)
````

#### Questão 2-5: busca binária como função de alta ordem
Usando recursão, implemente a função `busca_binaria(lst, chave, seletor_chave, f_igual, f_maior)`.

Esta função usa busca binária para encontrar uma `chave` em `lst`. Se encontrar, ela retorna o elemento correspondente em `lst`. Como `lst` é uma lista de tipos arbitrários, `busca_binaria` usa a função `seletor_chave` para obter a chave de um elemento da lista; usa as funções `f_igual` e `f_maior` para comparar a chave dada com a chave do elemento, determinando assim se são iguais ou se a primeira é maior que a segunda. Abaixo, temos exemplos de uso da função:

```python
from string import lower

# primeiro, um exemplo simples
print (busca_binaria([10, 15, 17, 19, 23, 31, 37, 43, 67, 81, 94], 19)) # 19

# agora, um exemplo mais complexo com tuplas como elementos; 
# chave em ordem decrescente e mix entre minusculas e maiusculas
dados = [(10, 'Tiago'), (15, 'Renato'), (17, 'Paula'), 
         (19, 'marcelo'), (23, 'joao'), (31, 'nubia'), 
         (37, 'Keila'), (43, 'Eduardo'), (67, 'Denys'), 
         (81, 'Beatriz'), (94, 'ana')]
print (busca_binaria(dados, 'ana', seletor_chave = lambda x: lower(x[1]),
                    f_maior = lambda x, y: x < y)) # (94, 'ana')
```

````
def busca_binaria(lst, chave,
                  seletor_chave = lambda x: x,
                  f_igual = lambda x, y: x == y,
                  f_maior = lambda x, y: x > y):
  if not lst:
    return None

  meio = len(lst) // 2
  e = seletor_chave(lst[meio])

  return lst[meio] if f_igual(e, chave)  \
    else busca_binaria(lst[:meio] if f_maior(e, chave) else lst[meio + 1:],
                       chave, seletor_chave, f_igual, f_maior)



## --------------------------------------- 
print( busca_binaria([10, 15, 17, 19, 23, 31, 37, 43, 67, 81, 94], 19) ) ## 19

dados = [(10, 'Tiago'),   (15, 'Renato'),  (17, 'Paula'), 
         (19, 'marcelo'), (23, 'joao'),    (31, 'nubia'), 
         (37, 'Keila'),   (43, 'Eduardo'), (67, 'Denys'), 
         (81, 'Beatriz'), (94, 'ana')]

from string import lower
print( busca_binaria(dados, 'ana',
                     seletor_chave=lambda x: lower(x[1]),
                     f_maior=lambda x, y: x < y) ) ## (94, 'ana')
````

### Um Interpretador $\mu$Forth

Nesta série de exercícios, vamos implementar um interpretador para a linguagem $\mu$Forth, usando Python funcional.

Como Forth, $\mu$Forth é uma linguagem baseada em pilha. Um programa em $\mu$Forth é uma sequência de instruções que modificam uma pilha. Cada instrução se refere a uma palavra previamente definida ou a definição de uma nova palavra. Por exemplo, no programa abaixo:

```
3 4 +
```

há três palavras, os números 3 e 4 e o sinal +. Os números 3 e 4 são inseridos na pilha. A palavra + indica que os dois números no topo da pilha (4, 3; 4 no topo) serão desempilhados, somados (+) e o resultado (7) será inserido na pilha. Assim, dada uma pilha vazia, após a execução de “3 4 +”, ela passa a ter um elemento, 7. A seguir, um exemplo mais complexo:

```
: quadrado dup * ;
3 quadrado
```

Na primeira linha é definida a palavra quadrado. Definições são iniciadas com um ‘:’ (dois-pontos). A palavra definida será substituída por todas as palavras até um ‘;’ (ponto-e-vírgula). No exemplo, a definição de quadrado duplica o valor no topo da pilha (palavra reservada `dup`) e então multiplica os dois elementos do topo (palavra `*`). Ou seja, após a execução de quadrado, o valor que estava no topo da pilha é substituído por seu quadrado. Na segunda linha, o valor 3 é inserido na pilha e a palavra quadrado é aplicada, de forma que o valor 3 é substituído por 9. 

````
####################################################################################
####################################### extra ######################################
####################################################################################

def test(fn, *args, **kwargs):
  """
  Para realizar testes unitários.
  Recebe uma função e seus argumentos e
  retorna uma função que executará a passada esperando uma saída desejada.
  Em caso de uma saída diferente da esperada, um print indicará a saída obtida.
  """
  def execute_waiting_this_output(expected_output):
    output = fn(*args, **kwargs)
    if (output != expected_output):
      print( '\x1b[31;1m✗\x1b[;1m {}({})\x1b[0m →\x1b[;1m {}\x1b[0m'.format(fn.__name__, ', '.join(map(str, args)), output) )

  return execute_waiting_this_output
````

#### Questão 3-1: uma pilha funcional
Vamos iniciar nossa implementação criando funções de manipulação de pilha  `empty` (verifica se pilha está vazia),  `push` (insere elemento no topo da pilha),  `pop` (retira o elemento no topo da pilha) e `top` (retorna o valor no topo da pilha), em estilo funcional. Ou seja, as funções recebem a pilha e a retornam modificada, de acordo com a operação executada. Veja os exemplos abaixo de uso: 

```python
ex_stack = [10, 20, 30, 40] # 40 é o topo da pilha
print(empty(ex_stack)) # False
print(push(ex_stack, 50)) # [10, 20, 30, 40, 50]
print(pop(ex_stack)) # [10, 20, 30]
print(top(ex_stack)) # 40
```

````
def empty(s):
  return len(s) == 0

def push(stack, e):
  return stack + [e]

def pop(stack):
  return stack[:-1]

def top(stack):
  return stack[-1] if not empty(stack) else None



## ---------------------------------------
ex_stack = [10, 20, 30, 40]

test(empty, ex_stack)(False)
test(push, ex_stack, 50)([10, 20, 30, 40, 50])
test(pop, ex_stack)([10, 20, 30])
test(top, ex_stack)(40)
````

#### Questão 3-5: roll
Crie a função `roll(s)` que opera sobre uma pilha `s`. Esta função rotaciona o $i$-ésimo elemento de uma pilha para o seu topo. Para tanto, ela inicialmente considera que o valor no topo da pilha corresponde à posição $i$ do restante da pilha. Ela então move o $i$-ésimo elemento para o topo. Implemente esta função usando as funções de operação em pilha que você implementou antes. Veja o exemplo a seguir: 

```python
# No exemplo abaixo, 2 corresponde à 3a posicao da pilha sem o 2, ou seja, a posição do 10
print(roll([10, 20, 30, 2])) # [20, 30, 10]
```

````
def roll(s):
  if empty(s):
    return []
  if top(s) >= len(s) - 1: ## guard clasue para evitar exceções
    return s[:]

  def split_until_0(i, sin, sout):
    ## Desempilha `sin` em `sout` enquanto `i` for maior que 1
    return (sin, sout) if i < 1 \
      else split_until_0( i - 1, pop(sin), push(sout, top(sin)) )

  def repush(sin, sout):
    ## Desempilha os elementos de `sin`, empilhando-os em `sout`
    return sout if empty(sin) \
      else repush( pop(sin), push(sout, top(sin)) )

  sin, sout = split_until_0(top(s), pop(s), [])
  return push( repush(sout, pop(sin)), top(sin) )



## ---------------------------------------
test(roll, [10, 20, 30, 2])([20, 30, 10])
test(roll, [10, 20, 30, 1])([10, 30, 20])
test(roll, [10, 20, 30, 3])([10, 20, 30, 3])
test(roll, [10, 20, 30, 40, 3])([20, 30, 40, 10])
test(roll, [13, 12, 11, 10, 2])([13, 11, 10, 12])
test(roll, [7, 6, 5, 1])([7, 5, 6])
test(roll, [8, 7, 5, 6, 2])([8, 5, 6, 7])
````

#### Questão 3-9: comparação de linguages
Baseado em sua experiência com Python Funcional, a compare com Forth, considerando os seguintes critérios:

Apesar da **simplicidade** global em ambos, o maior grau de **ortogonalidade** do Forth é mais simples que o python funcional, apesar de fornecer um número maior de componentes básicos.

O Forth não possui uma facilidade para definir **tipos de dados**, visto seu alto grau de ortogonalidade. De modo que, por exemplo, um booleano é representado pela convenção númerica. O mesmo não acontece no Python, melhorando a legibilidade.

Em relação à **sintaxe**, apesar da simplicidade adotada pelo Forth, a legibilidade é prejudicada. Ao usar grupos de palavras reservadas como delimitadores de instruções de controle, o número de palavras reservadas aumenta. No Python funcional, muitas instruções indicam sua funcionalidade o que não ocorre no Forth, só que, por exemplo, "r>" tem um significado implícito nada intuitivo. A indentação como definição de escopo não prejudica a legibilidade se as funções forem simples (e curtas em linhas), como o esperado. Já a ausência dessa definição no Forth, a torna mais "aceitável" e fácil de entender, já que as instruções são cadeias de palavras e somente isso.

A **capacidade de abstração** do processo em Forth, é simples e consistente, muito devido à ortogonalidade facilitando a escrita. Porém, seria necessário vários níveis de abstrações para definir uma estrutura mais complexa, como uma árvore binária. O mesmo problema ocorre no Python funcional, porém de forma menos complexa de lidar.

O Python funcional tem geradores que permitem que várias computações possam ser realizadas em um programa. O Forth também tem essa **expressividade**, porém, a capacidade de escrita é mais prejudicada.

Como o Forth trabalha com pilhas de números, a **checagem de tipos** se torna irrelevante. Enquanto no Python funcional a verificação ocorre de maneira indepentendete, mas, ainda pode ser reforçada usando-se funções que acessam meta informações de um elemento. 

Pyhton funcional usa um mecanismo para **manipular exceções** que intercepta erros em tempo de execução, porém a confiabilidade se reduz à isto. Já o Forth não oferece tal suporte, assumindo que tudo deve ocorrer de maneira esperada.
